## Check If the Account is Balance

```python
print(f"Total Credit: {df['DR/CR'].sum()}")
```

## Revenue

```python
New_Revenue=['Revenue','Other Income']
Account_Revenue = df[df['FINANCIAL STATEMENT LINE'].isin(New_Revenue)].copy()
Total_Revenue=Account_Revenue['DR/CR'].sum()
print(f'Revenue Statement: {Total_Revenue:,.1f}')
```

## Expenses

```python
New_Expenses2=['Cost of Sales','Distribution Expenses','Depreciation Charged','Administrative Expenses','Finance Charges']
Account_Expenses2 = df[df['FINANCIAL STATEMENT LINE'].isin(New_Expenses2)].copy()
Total_Expenses=Account_Expenses2['DR/CR'].sum()
print(f'Expenses_statement: {Total_Expenses:,.1f}')
```

## Profit and Loss statements

```python
PL = ['Other Gains/Lossess']
PL_Account = df[df['FINANCIAL STATEMENT LINE'].isin( PL)].copy()
PL_Account.head()
```

```python
PL_statement= PL_Account['DR/CR'].sum()
print(f'Total PL: {PL_statement:,.1f}')
```

```python
# Profit or Loss
profit_or_loss = Total_Revenue - Total_Expenses

print(f"Total Revenue: {Total_Revenue:,.1f}")
print(f"Total Expenses: {Total_Expenses:,.1f}")

print("\n=== PROFIT AND LOSS STATEMENT ===")
print(f"Profit And Loss : {profit_or_loss:,.1f}")

if profit_or_loss > 0:
    print(f"Net Profit: {profit_or_loss:,.1f}")
else:
    print(f"Net Loss: {profit_or_loss:,.1f}")
```

```python

Gain_ex=['Exchange Rate (Gain)','Exchange Rate Loss']
Gain=df[df['DESCRIPTION'].isin(Gain_ex)].copy()
Gain_statement=Gain['DR/CR'].sum()
print(f'Gain_statemen: {Gain_statement:,.1f}')
Expenses_statement2=Gain_statement + Total_Expenses
print(f'Expenses_statement: {Expenses_statement2:,.1f}')
```

## Net Profit

```python
Net_Profit=Total_Revenue-Total_Expenses
print(f'Net_Profit: {Net_Profit:,.1f}')
```

## Add depreciation

```python
Dep=['Depreciation Expenses']
Account_Dep = df[df['DESCRIPTION'].isin(Dep)].copy()
Depreciation_statement=Account_Dep['DR/CR'].sum()

print(f'depreciation: {Depreciation_statement:,.1f}')
```

## Donations

```python
Don=['Corporate and Sallah Gift','Donations','Entertainment','Zakkat']
Account_Don= df[df['DESCRIPTION'].isin(Don)].copy()
Donations_statement=Account_Don['DR/CR'].sum()
print(f'Donations_statement: {Donations_statement:,.1f}')
```

### PPE

```python
Ann=['Tools','Plant & Machinery','Office Partitioning','Motor Vehicles','Fittings & Equipment','Building','Land ']
Account_Ann= df[df['DESCRIPTION'].isin(Ann)].copy()
Annual_All=Account_Ann['DR/CR'].sum()
print(f'PPE: {Annual_All:,.1f}')
```

```python
Initial_Allowance=Annual_All/ 2
Allowance = round(Initial_Allowance, 1) 
print(f'Initial Allowance: {Allowance:,.1f}')
```

## Annual Allowance

```python
Annual_Allowance3=Annual_All-Allowance / 4
print(f'Annual Allowance: {Annual_Allowance3:,.1f}')
```

## Adjusted/Assesable Profit

```python
Adjusted_Assesable_Profit= Net_Profit - Depreciation_statement - Donations_statement
print(f'Adjusted_Assesable_Profit: {Adjusted_Assesable_Profit:,.1f}')
```

## Profit

```python
Profit=Adjusted_Assesable_Profit + Annual_Allowance3
print(f"profit: {Profit:,.1f}")
```

## Tax

```python
Tax=Profit * 0.30
print(f'Tax: {Tax:,.1f}')
```
